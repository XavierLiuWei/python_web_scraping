```python
from tmdbv3api import TMDb
from tmdbv3api import Movie
import pandas as pd
import requests
import json

# 輸入電影id範圍
max_id=1095005
min_id=1094506

api_key = '740de416f4eeb91d5be250fdbfd226e8'

# 建議個空的list
movie_ids=[]

# # ng_list保留作預備用
# movie_ng_ids=[]

while True:
    # API請求獲取電影基本資訊
    url=f'https://api.themoviedb.org/3/movie/{max_id}?api_key={api_key}&region=US&language=en-US'
    response = requests.get(url)

    # 資訊獲取
    data = json.loads(response.text)
    
    # 新增到list內,
    movie_ids.append(max_id) if 'id' in data.keys()
    print(f'current movie: {max_id}')
    
    max_id -=1
    if max_id < min_id:
        break
        
#將電影id存成json檔
json_str = json.dumps(movie_ids)
 
with open(f'movie_id_{min_id}~{max_id}.json', 'w') as f:
    f.write(json_str)
```
